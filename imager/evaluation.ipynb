# Requirements

```python
from testing_classifier import thread_mapReduce
import os
import glob
import pandas as pd
import h5py
import numpy as np
```

# Read CSV

```python
path = os.getcwd() + "/../Datasets"
csv_files = glob.glob(os.path.join(path, "*.csv"))
csv_dfs = {}
for f in csv_files:
    file_name = f.split("\\")[-1]

    # remove the .csv extension
    file_name = file_name[:-4]

    if file_name == "Amazon-Products":
        continue

    # read the csv file 
    df = pd.read_csv(f)
    csv_dfs[file_name] = df
```

# Read embeddings

```python
def get_embeddings(file_name):
    folder_path = 'images'
    with h5py.File(folder_path + f"/{file_name}.hdf5", 'r') as hdf:
        if "images" in hdf:
            embeddings = hdf["images"][:]
        else:
            embeddings = np.zeros((0, 512))
        return embeddings.copy()
```

```python
urls = {file: csv_dfs[file]["image"] for file in csv_dfs}
embeddings = {file: get_embeddings(file) for file in csv_dfs}
```

### testing

```python
file_names = [line.strip() for line in open("file_names.txt", "r")]
results = thread_mapReduce(file_names, emb, top_k=10)
```

```python
df = pd.DataFrame(columns=file_names)
for csv_result in results:
    csv_name = csv_result[0]
    scores = csv_result[1].flatten()
    indices = csv_result[2].flatten()
    # merge scores and indices to tuple
    maxes = list(zip(indices, scores))
    df[csv_name] = maxes
```

```python
import h5py
import numpy as np
import matplotlib.pyplot as plt
import requests
from io import BytesIO
from PIL import Image
from testing_classifier import get_embedding_links, get_cosine_similarities
```

```python
# Display the top 10 closest images
fig, axes = plt.subplots(2, 5, figsize=(15, 6))
axes = axes.flatten()

# input_img_path = "evaluation_images/bottle.png"
# input_img = load_image(input_img_path)
# input_embedding = get_image_embeddings(input_img)
# np.save("bottle.npy", input_embedding)

query_product_embedding = np.load('bottle.npy')
all_embeddings, image_links = get_embedding_links()
similarities, top_10_indices = get_cosine_similarities(query_product_embedding, all_embeddings, k=10)

for ax, idx in zip(axes, top_10_indices):
    image_link = image_links[idx]
    if image_link:
        response = requests.get(image_link)
        img = Image.open(BytesIO(response.content))
        ax.imshow(img)
    else:
        ax.text(0.5, 0.5, 'No Image', horizontalalignment='center', verticalalignment='center')
    ax.axis('off')
    ax.set_title(f'Similarity: {similarities[idx]:.4f}')

plt.tight_layout()
plt.show()
```
