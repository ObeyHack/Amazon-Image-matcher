# Requirements

```python
from imager.vgg.VGG import get_image_embeddings, get_similarity_score, load_image, get_score
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

# Load Image

```python
def show_image(image_path):
  image = mpimg.imread(image_path)
  imgplot = plt.imshow(image)
  plt.show()
```

```python
crown = '../sunflower/crown.jpg'
king = '../sunflower/king.jpg'
man = '../sunflower/man.jpg'
image1 = load_image(crown)
image2 = load_image(king)
image3 = load_image(man)
```

# Image 1

```python
show_image(crown)
```

# Image 2

```python
show_image(king)
```

```python
show_image(man)
```

# Image Embeddings

```python
image1_embedding = get_image_embeddings(image1)
print(image1_embedding.shape)
print(image1_embedding)
```

```python
image2_embedding = get_image_embeddings(image2)
print(image2_embedding.shape)
print(image2_embedding)
```

```python
image3_embedding = get_image_embeddings(image3)
print(image3_embedding.shape)
print(image3_embedding)
```

```python
king_without_crown_emb = image2_embedding - image1_embedding
print(king_without_crown_emb.shape)
print(king_without_crown_emb)
```

```python
get_score(king_without_crown_emb, image3_embedding)
```
