# Evaluation

```python
import os
import h5py
import numpy as np
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
import matplotlib.pyplot as plt
import requests
from io import BytesIO
from PIL import Image

# Path to the folder containing HDF5 files
folder_path = 'images'
# Path to the folder containing CSV files with image links
csv_folder_path = r'..\Datasets'

# Initialize an empty list to store embeddings and image links
all_embeddings = []
image_links = []

def list_datasets(file_path):
    """List all dataset names in an HDF5 file."""
    with h5py.File(file_path, 'r') as hdf:
        return [name for name in hdf.keys()]

# Iterate through each file in the folder
for file_name in os.listdir(folder_path):
    # if the hdf5 file is dataset.hdf5, continue
    if file_name == 'dataset.hdf5':  # TODO: why does it exist?
        continue
    if file_name.endswith('.hdf5'):
        file_path = os.path.join(folder_path, file_name)
        csv_file_path = os.path.join(csv_folder_path, file_name.replace('.hdf5', '.csv'))

        dataset_names = list_datasets(file_path)

        with h5py.File(file_path, 'r') as hdf:
            # Read the corresponding CSV file to get image links
            df = pd.read_csv(csv_file_path)
            for dataset_name in dataset_names:
                embeddings = hdf[dataset_name][:]
                if embeddings.size > 0:
                    # Detect the actual embedding size
                    actual_embedding_size = embeddings.shape[-1]
                    for i, embedding in enumerate(embeddings):
                        all_embeddings.append(embedding)
                        image_links.append(df['image'].iloc[i])
                else:
                    all_embeddings.append(np.zeros(actual_embedding_size))  # Handle empty embeddings case
                    image_links.append(None)  # Handle empty embeddings case

# Convert lists to NumPy arrays
all_embeddings = np.array(all_embeddings)
image_links = np.array(image_links)

# Ensure the embeddings array is 2-dimensional
all_embeddings = all_embeddings.reshape(all_embeddings.shape[0], -1)

# Save the concatenated embeddings and image links as NumPy arrays
np.save('all_embeddings.npy', all_embeddings)
np.save('image_links.npy', image_links)

print(f'All embeddings saved as a NumPy array with shape: {all_embeddings.shape}')
print(f'Image links saved as a NumPy array with shape: {image_links.shape}')

# Create a query embedding (using the first embedding as an example)
query_product_embedding = all_embeddings[100]

# Calculate cosine similarities
similarities = cosine_similarity(query_product_embedding.reshape(1, -1), all_embeddings).flatten()

# Get indices of top 10 closest embeddings
top_10_indices = np.argsort(-similarities)[:10]

# Display the top 10 closest images
fig, axes = plt.subplots(2, 5, figsize=(15, 6))
axes = axes.flatten()

for ax, idx in zip(axes, top_10_indices):
    image_link = image_links[idx]
    if image_link:
        response = requests.get(image_link)
        img = Image.open(BytesIO(response.content))
        ax.imshow(img)
    else:
        ax.text(0.5, 0.5, 'No Image', horizontalalignment='center', verticalalignment='center')
    ax.axis('off')
    ax.set_title(f'Similarity: {similarities[idx]:.4f}')

plt.tight_layout()
plt.show()
```
